Что делает строка целиком:
Она извлекает числовое значение количества памяти из строкового описания и создаёт новый столбец Memory_Amount.

Пошаговая расшифровка:
1. laptop['Memory']
Обращение к столбцу Memory в DataFrame laptop

Пример данных: ['8 GB', '16 GB DDR4', '512 GB SSD']

2. .apply(lambda i: ...)
Применяет функцию к КАЖДОЙ ячейке столбца Memory

lambda i: — анонимная функция, где i это одна ячейка

3. i.split(' ')
Разбивает строку по пробелам

Пример: '16 GB DDR4'.split(' ') → ['16', 'GB', 'DDR4']

4. [0]
Берёт первый элемент из получившегося списка

Пример: ['16', 'GB', 'DDR4'][0] → '16'

5. laptop['Memory_Amount'] = ...
Создаёт новый столбец Memory_Amount с результатами

```python
def convert_to_gb(x):
    memory=x.split(' ')[0]
    if memory.endswith('GB'):
        return int(memory.replace('GB' , ''))
    else:
        return int(float(memory.replace('TB' , ''))*1024)
laptop['Memory_Amount1'] = laptop['Memory'].apply(convert_to_gb)
# laptop['Memory_Amount'] = laptop['Memory'].apply(lambda i:''.join(filter(str.isdigit, i))) #без перевода в гиги
laptop['Memory_Amount2'] = laptop['Memory'].apply(lambda i:int(i.split(' ')[0].replace('GB', '')) if 'GB' in i else int(float(i.split(' ')[0].replace('TB', '')) * 1024))
laptop.head(20)
```

## Задание 2.
2.1 Создайте признак SSD, который изначально равен 0
2.2 Поставьте в признаке SSD 1, если ноутбук
действительно с типом носителя SSD
2.3 Уберите в признаке Weight значения 'kg' и поменяйте его

```python
# Создаем признак SSD изначально равный 0
laptop['SSD'] = 0

# Вариант 1: Простая логика - если есть SSD в любом виде
laptop['SSD'] = laptop['Memory'].str.contains('SSD', case=False, na=False).astype(int)
laptop.head()
```

```python
# ИЛИ Вариант 2: Более точная логика с учетом комбинаций
def check_ssd(memory_text):
    memory_str = str(memory_text).upper()
    if 'SSD' in memory_str:
        return 1
    return 0

laptop['SSD1'] = laptop['Memory'].apply(check_ssd)
laptop.head()
```

```python
# Создадим несколько признаков для разных типов накопителей
laptop['Storage_SSD'] = laptop['Memory'].str.contains('SSD', case=False, na=False).astype(int)
laptop['Storage_HDD'] = laptop['Memory'].str.contains('HDD', case=False, na=False).astype(int)
laptop['Storage_Flash'] = laptop['Memory'].str.contains('Flash', case=False, na=False).astype(int)
laptop['Storage_Hybrid'] = laptop['Memory'].str.contains('Hybrid', case=False, na=False).astype(int)

# Посмотрим распределение
print("SSD:", laptop['Storage_SSD'].sum())
print("HDD:", laptop['Storage_HDD'].sum()) 
print("Flash:", laptop['Storage_Flash'].sum())
print("Hybrid:", laptop['Storage_Hybrid'].sum())
```

```python
# Количество типов накопителей в одном ноутбуке
laptop['Storage_Types_Count'] = (laptop['Storage_SSD'] + 
                                laptop['Storage_HDD'] + 
                                laptop['Storage_Flash'] + 
                                laptop['Storage_Hybrid'])

# Проверим комбинации
print(laptop['Storage_Types_Count'].value_counts())
```

```python
# Посмотрим примеры с разными типами накопителей
sample_data = laptop[['Memory', 'Storage_SSD', 'Storage_HDD', 'Storage_Flash', 'Storage_Hybrid']]
print(sample_data.head())
```

```python
def convert_to_num(x):
    mem1 = x.split('+')
    if len(mem1) > 1:
        mem = 0
        for m in mem1:
            mem2 = m.split()[0]
            if mem2.endswith('GB'):
                mem += int(mem2.replace('GB', ''))
            else:
                mem += int(float(mem2.replace('TB', ''))*1024)
        return mem
    else:
        mem2 = mem1[0].split()[0]
        if mem2.endswith('GB'):
            return int(mem2.replace('GB', ''))
        else:
            return int(float(mem2.replace('TB', ''))*1024)
laptop['SSD1'] = laptop['Memory'].apply(convert_to_num)
laptop.head(1000)
```
