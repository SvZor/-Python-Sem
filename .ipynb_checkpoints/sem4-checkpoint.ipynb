```python
import numpy as np
# Создаем фигуру для диаграммы с заданным размером
plt.figure(figsize=(8, 8))

# Получаем распределение типов носителей (Storage_Type)
# value_counts() подсчитывает количество каждого уникального значения в столбце
storage_counts = laptop['Memory_Type'].value_counts()

# Определяем цвета для каждого сегмента диаграммы
# Можно использовать готовую палитру или задать свои цвета
colors = plt.cm.Set3(np.linspace(0, 1, len(storage_counts)))

# Создаем круговую диаграмму (pie chart)
# storage_counts.values - числовые значения (количество каждого типа)
# labels=storage_counts.index - подписи для каждого сегмента (названия типов носителей)
# autopct='%1.1f%%' - отображает проценты с одним знаком после запятой
# startangle=90 - начальный угол поворота (первый сегмент начинается сверху)
# colors=colors - применяем заданную цветовую схему
# shadow=True - добавляет тень для объемного эффекта
plt.pie(storage_counts.values, 
        labels=storage_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=colors, 
        shadow=True)

# Устанавливаем заголовок диаграммы
plt.title('Распределение типов носителей памяти', fontsize=16, fontweight='bold')

# Добавляем легенду для лучшей читаемости
# loc='best' - автоматически выбирает лучшее расположение
# bbox_to_anchor=(1, 0.5) - смещает легенду вправо от диаграммы
plt.legend(storage_counts.index, 
           title="Типы носителей",
           loc='center left', 
           bbox_to_anchor=(1, 0.5))

# Устанавливаем равное соотношение осей (чтобы диаграмма была круглой, а не овальной)
plt.axis('equal')

# Автоматически регулирует отступы, чтобы все элементы поместились
plt.tight_layout()

# Отображаем диаграмму
plt.show()
```

```python
# Альтернативный вариант с выносом малых сегментов
plt.figure(figsize=(6, 6))

# Определяем порог для выноса малых сегментов (например, менее 5%)
threshold = 5
total = storage_counts.sum()

# Создаем маску для сегментов меньше порога
mask = (storage_counts / total * 100) < threshold

# Для вынесенных сегментов объединяем в "Другие"
if mask.any():
    others_sum = storage_counts[mask].sum()
    main_counts = storage_counts[~mask]
    main_counts['Другие'] = others_sum
else:
    main_counts = storage_counts

# Создаем диаграмму с выносом малых сегментов
plt.pie(main_counts.values, 
        labels=main_counts.index, 
        autopct='%1.1f%%',
        explode=[0.1 if i == main_counts.values.argmax() else 0 for i in range(len(main_counts))],  # Выделяем самый большой сегмент
        startangle=90,
        colors=plt.cm.Pastel1(np.linspace(0, 1, len(main_counts))),
        shadow=True)

plt.title('Распределение типов носителей памяти\n(малые сегменты объединены в "Другие")', 
          fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Горизонтальная бар-диаграмма для сравнения количеств
plt.figure(figsize=(8, 4))

# Сортируем по убыванию для лучшей читаемости
storage_sorted = storage_counts.sort_values(ascending=True)

# Создаем горизонтальную бар-диаграмму
plt.barh(storage_sorted.index, storage_sorted.values, color='skyblue')

# Добавляем значения на столбцы
for i, value in enumerate(storage_sorted.values):
    plt.text(value + 0.1, i, f'{value}', va='center')

plt.xlabel('Количество ноутбуков')
plt.ylabel('Тип носителя')
plt.title('Распределение типов носителей памяти', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Функция для сохранения графика с выбором места
def save_plot_interactive():
    # Создаем скрытое окно tkinter (оно не будет видно)
    root = tk.Tk()
    root.withdraw()  # Скрываем основное окно
    
    # Открываем диалоговое окно для выбора места сохранения
    # title - заголовок окна
    # defaultextension - расширение файла по умолчанию
    # filetypes - допустимые типы файлов
    file_path = filedialog.asksaveasfilename(
        title="Сохранить график как...",
        defaultextension=".png",
        filetypes=[
            ("PNG files", "*.png"),
            ("JPEG files", "*.jpg"),
            ("PDF files", "*.pdf"),
            ("SVG files", "*.svg"),
            ("All files", "*.*")
        ]
    )
    
    # Если пользователь выбрал путь (не нажал "Отмена")
    if file_path:
        # Сохраняем график с высоким качеством
        # dpi - разрешение (точек на дюйм)
        # bbox_inches='tight' - обрезает пустые поля вокруг графика
        plt.savefig(file_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"График успешно сохранен как: {file_path}")
    else:
        print("Сохранение отменено")
    
    # Закрываем окно tkinter
    root.destroy()

# Отображаем график
plt.show()

# После закрытия окна с графиком предлагаем сохранить
# Спрашиваем пользователя, хочет ли он сохранить график
save_choice = input("Хотите сохранить график? (y/n): ").lower().strip()

if save_choice in ['y', 'yes', 'д', 'да']:
    save_plot_interactive()
else:
    print("График не сохранен")

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Создаем копию датафрейма для преобразований
laptop_clean = laptop.copy()

# 1. Преобразуем столбец Ram из формата '8GB' в числовой
print("Уникальные значения в столбце Ram до очистки:")
print(laptop_clean['Ram'].unique())

# Удаляем 'GB' и преобразуем в числа
laptop_clean['Ram_numeric'] = laptop_clean['Ram'].str.replace('GB', '').astype(int)

print("\nУникальные значения в столбце Ram после очистки:")
print(laptop_clean['Ram_numeric'].unique())

# 2. Преобразование категориальных переменных в числовые с помощью pandas
categorical_columns = ['Cpu_Company', 'Memory_Type', 'Company']

for col in categorical_columns:
    # Создаем числовые коды для категориальных переменных
    laptop_clean[col + '_encoded'] = laptop_clean[col].astype('category').cat.codes
    print(f"\nКодирование для {col}:")
    encoding_map = dict(zip(laptop_clean[col], laptop_clean[col + '_encoded']))
    for category, code in encoding_map.items():
        print(f"  {category} -> {code}")

# 3. Матрица корреляции для числовых характеристик
numeric_columns = ['Price_euros', 'Ram_numeric', 'Weight', 'Inches']
plt.figure(figsize=(8, 6))
correlation_matrix = laptop_clean[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', linewidths=0.5)
plt.title('Матрица корреляции числовых характеристик')
plt.tight_layout()
plt.show()

# 4. Корреляция Price_euros с другими числовыми переменными
price_corr = laptop_clean[numeric_columns].corr()['Price_euros'].sort_values(ascending=False)
print("Корреляция цены с другими характеристиками:")
print(price_corr)

# 5. Heatmap корреляции с Price_euros
plt.figure(figsize=(6, 4))
price_corr_df = pd.DataFrame(price_corr[1:])  # Исключаем сам Price_euros
sns.heatmap(price_corr_df, annot=True, cmap='RdYlBu', center=0, 
            square=True, fmt='.2f', linewidths=0.5)
plt.title('Корреляция характеристик с Price_euros')
plt.tight_layout()
plt.show()

# 6. Расширенная матрица корреляции с закодированными категориальными переменными
extended_columns = ['Price_euros', 'Ram_numeric', 'Weight', 'Inches'] + \
                  ['Cpu_Company_encoded', 'Memory_Type_encoded', 'Company_encoded']

plt.figure(figsize=(10, 8))
extended_corr = laptop_clean[extended_columns].corr()
sns.heatmap(extended_corr, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f', linewidths=0.5)
plt.title('Расширенная матрица корреляции\n(включая закодированные категориальные переменные)')
plt.tight_layout()
plt.show()

# 7. Корреляция Price_euros со всеми переменными (включая закодированные)
extended_price_corr = laptop_clean[extended_columns].corr()['Price_euros'].sort_values(ascending=False)
print("\nРасширенная корреляция Price_euros со всеми переменными:")
print(extended_price_corr)

# 8. Визуализация топ-корреляций с ценой
top_corr = extended_price_corr[1:7]  # Топ-6 кроме самой цены
plt.figure(figsize=(10, 6))
sns.barplot(x=top_corr.values, y=top_corr.index, palette='viridis')
plt.xlabel('Коэффициент корреляции')
plt.ylabel('Характеристики')
plt.title('Топ-6 характеристик по корреляции с Price_euros')
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.tight_layout()
plt.show()

# 9. Pairplot для визуализации взаимосвязей
top_corr_features = ['Price_euros'] + top_corr.index[:3].tolist()
# Переименовываем для читаемости на графике
rename_dict = {'Ram_numeric': 'RAM (GB)'}
top_corr_features_renamed = [rename_dict.get(x, x) for x in top_corr_features]
laptop_plot = laptop_clean[top_corr_features].rename(columns=rename_dict)

sns.pairplot(laptop_plot, diag_kind='hist', corner=True)
plt.suptitle('Pairplot наиболее коррелирующих переменных', y=1.02)
plt.show()

# 10. Дополнительно: Boxplot с очищенными данными RAM
plt.figure(figsize=(12, 6))
sns.boxplot(x=laptop_clean['Price_euros'], y=laptop_clean['Ram_numeric'], whis=1.5)
plt.xlabel('Price_euros')
plt.ylabel('RAM (GB)')
plt.title('Distribution of Price_euros by RAM (numeric)')
plt.show()

# 11. Анализ категориальных переменных через группировку
print("\nСредняя цена по производителям CPU:")
cpu_price_stats = laptop_clean.groupby('Cpu_Company')['Price_euros'].agg(['mean', 'median', 'count'])
print(cpu_price_stats.sort_values('mean', ascending=False))

print("\nСредняя цена по типам памяти:")
memory_price_stats = laptop_clean.groupby('Memory_Type')['Price_euros'].agg(['mean', 'median', 'count'])
print(memory_price_stats.sort_values('mean', ascending=False))

print("\nСредняя цена по компаниям-производителям:")
company_price_stats = laptop_clean.groupby('Company')['Price_euros'].agg(['mean', 'median', 'count'])
print(company_price_stats.sort_values('mean', ascending=False))
```
