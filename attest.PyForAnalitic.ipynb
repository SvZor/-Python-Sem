```python
# Создаем сегменты по площади
bins = [0, 1000, 1500, 2000, 2500, 3000, 4000, 10000]
labels = ['<1000', '1000-1500', '1500-2000', '2000-2500', '2500-3000', '3000-4000', '>4000']

sqft_segments = pd.cut(kc_house['sqft_living'], bins=bins, labels=labels)
segment_stats = sqft_segments.value_counts().sort_index()

print("\nРАСПРЕДЕЛЕНИЕ ПО СЕГМЕНТАМ ПЛОЩАДИ:")
print("=" * 40)
for segment, count in segment_stats.items():
    percentage = (count / len(kc_house)) * 100
    print(f"{segment} кв.футов: {count} домов ({percentage:.1f}%)")
```

```python
print("\nКЛЮЧЕВЫЕ ВЫВОДЫ ПО ЖИЛОЙ ПЛОЩАДИ:")
print("=" * 45)
print(f"1. ТИПИЧНЫЙ РАЗМЕР: {sqft_stats['median']:.0f} кв.футов (медиана)")
print(f"2. ОСНОВНОЙ РЫНОК: {sqft_stats['q1']:.0f} - {sqft_stats['q3']:.0f} кв.футов (50% домов)")
print(f"3. ДИАПАЗОН: от {sqft_stats['min']:.0f} до {sqft_stats['max']:.0f} кв.футов")

# Анализ асимметрии
if sqft_stats['skew'] > 0.5:
    print("4. РАСПРЕДЕЛЕНИЕ: правостороннее (есть дома с большой площадью)")
elif sqft_stats['skew'] < -0.5:
    print("4. РАСПРЕДЕЛЕНИЕ: левостороннее (преобладают небольшие дома)")
else:
    print("4. РАСПРЕДЕЛЕНИЕ: близко к нормальному")

# Анализ компактности
cv = (sqft_stats['std'] / sqft_stats['mean']) * 100
print(f"5. ВАРИАБЕЛЬНОСТЬ: {cv:.1f}% (коэффициент вариации)")

# Самый популярный сегмент
most_common_segment = segment_stats.idxmax()
most_common_count = segment_stats.max()
most_common_pct = (most_common_count / len(kc_house)) * 100
print(f"6. ПОПУЛЯРНЫЙ СЕГМЕНТ: {most_common_segment} кв.футов ({most_common_pct:.1f}% домов)")

# Анализ выбросов
large_houses = len(kc_house[kc_house['sqft_living'] > sqft_stats['q95']])
print(f"7. КРУПНЫЕ ДОМА: {large_houses} домов (> {sqft_stats['q95']:.0f} кв.футов)")

print("\nПЕРЕВОД В КВАДРАТНЫЕ МЕТРЫ:")
print("=" * 35)
print(f"Медианная площадь: {sqft_stats['median'] * 0.0929:.0f} кв.м")
print(f"Типичный диапазон: {sqft_stats['q1'] * 0.0929:.0f} - {sqft_stats['q3'] * 0.0929:.0f} кв.м")
print(f"Средняя площадь: {sqft_stats['mean'] * 0.0929:.0f} кв.м")
```

```python
# Анализ по временным периодам
def get_construction_period(year):
    if year < 1950:
        return 'До 1950'
    elif year < 1980:
        return '1950-1979'
    elif year < 2000:
        return '1980-1999'
    elif year < 2010:
        return '2000-2009'
    else:
        return '2010 и новее'

kc_house['period'] = kc_house['yr_built'].apply(get_construction_period)
period_stats = kc_house['period'].value_counts()

print("\nРАСПРЕДЕЛЕНИЕ ПО ПЕРИОДАМ ПОСТРОЙКИ:")
print("=" * 40)
for period, count in period_stats.items():
    percentage = (count / len(kc_house)) * 100
    print(f"{period}: {count} домов ({percentage:.1f}%)")
```

```python
# Статистики цен по периодам
price_by_period = kc_house.groupby('period').agg({
    'price': ['count', 'mean', 'median', 'min', 'max'],
    'yr_built': 'mean'
}).round(0)

print("\nЦЕНЫ ПО ПЕРИОДАМ ПОСТРОЙКИ:")
print("=" * 35)
for period in price_by_period.index:
    data = price_by_period.loc[period]
    print(f"\n{period}:")
    print(f"  Количество: {data[('price', 'count')]} домов")
    print(f"  Средняя цена: ${data[('price', 'mean')]:,.0f}")
    print(f"  Медианная цена: ${data[('price', 'median')]:,.0f}")
```

```python
# Корреляция года постройки с ценой и площадью
correlation_with_price = kc_house['yr_built'].corr(kc_house['price'])
correlation_with_sqft = kc_house['yr_built'].corr(kc_house['sqft_living'])

print(f"\nКОРРЕЛЯЦИОННЫЙ АНАЛИЗ:")
print("=" * 25)
print(f"Корреляция года постройки с ценой: {correlation_with_price:.3f}")
print(f"Корреляция года постройки с площадью: {correlation_with_sqft:.3f}")

# Интерпретация корреляции
if abs(correlation_with_price) > 0.5:
    strength = "сильная"
elif abs(correlation_with_price) > 0.3:
    strength = "умеренная"
elif abs(correlation_with_price) > 0.1:
    strength = "слабая"
else:
    strength = "очень слабая"

direction = "положительная" if correlation_with_price > 0 else "отрицательная"
print(f"Интерпретация: {strength} {direction} связь")
```

```python
# Вычисляем возраст домов
current_year = pd.Timestamp.now().year
kc_house['age'] = current_year - kc_house['yr_built']

print("\nКЛЮЧЕВЫЕ ВЫВОДЫ ПО ГОДУ ПОСТРОЙКИ:")
print("=" * 45)
print(f"1. ДИАПАЗОН ЛЕТ: {year_stats['min']} - {year_stats['max']}")
print(f"2. ТИПИЧНЫЙ ГОД: {year_stats['median']} (медиана)")
print(f"3. СРЕДНИЙ ВОЗРАСТ ДОМОВ: {kc_house['age'].mean():.1f} лет")
print(f"4. САМЫЙ РАСПРОСТРАНЕННЫЙ ПЕРИОД: {period_stats.index[0]}")
print(f"5. ВЛИЯНИЕ НА ЦЕНУ: {direction} корреляция ({correlation_with_price:.3f})")

# Анализ трендов
if correlation_with_price > 0.1:
    print("6. ТРЕНД: Новые дома обычно дороже старых")
elif correlation_with_price < -0.1:
    print("6. ТРЕНД: Старые дома обычно дороже новых")
else:
    print("6. ТРЕНД: Год постройки слабо влияет на цену")

# Анализ распределения
if year_stats['std'] > 30:
    print("7. РАЗНООБРАЗИЕ: Большой разброс по годам постройки")
else:
    print("7. РАЗНООБРАЗИЕ: Относительно однородный фонд недвижимости")

top_years = kc_house['yr_built'].value_counts().head(5)

print("\nТОП-5 ЛЕТ ПО КОЛИЧЕСТВУ ПОСТРОЕННЫХ ДОМОВ:")
print("=" * 50)
for year, count in top_years.items():
    percentage = (count / len(kc_house)) * 100
    print(f"{year} год: {count} домов ({percentage:.1f}%)")
```

```python
print("\n📊 КОЛИЧЕСТВЕННАЯ ОЦЕНКА ВЛИЯНИЯ ФАКТОРОВ:")
print("=" * 50)

# Корреляции с ценой
price_correlations = kc_house[['sqft_living', 'grade', 'bathrooms', 'waterfront', 'yr_built']].corrwith(kc_house['price']).sort_values(ascending=False)

for factor, corr in price_correlations.items():
    factor_name = {
        'sqft_living': 'Жилая площадь',
        'grade': 'Состояние дома',
        'bathrooms': 'Количество ванных',
        'waterfront': 'Вид на набережную',
        'yr_built': 'Год постройки'
    }[factor]
    
    strength = "ОЧЕНЬ СИЛЬНАЯ" if corr > 0.5 else "СИЛЬНАЯ" if corr > 0.3 else "УМЕРЕННАЯ" if corr > 0.1 else "СЛАБАЯ"
    print(f"• {factor_name:20} : {corr:.3f} ({strength})")

# Ценовые премии
print(f"\n💰 ЦЕНОВЫЕ ПРЕМИИ:")
waterfront_premium = (kc_house[kc_house['waterfront'] == 1]['price'].mean() / 
                     kc_house[kc_house['waterfront'] == 0]['price'].mean() - 1) * 100
print(f"• Вид на набережную: +{waterfront_premium:.0f}% к стоимости")

grade_premium = (kc_house[kc_house['grade'] >= 10]['price'].mean() / 
                kc_house[kc_house['grade'] <= 6]['price'].mean() - 1) * 100
print(f"• Высокое качество (10+ vs 6-): +{grade_premium:.0f}%")

print(f"\n🎯 ВЫВОД: Инвестируйте в улучшение качества и выбирайте объекты")
print("           с потенциалом добавленной стоимости через реновацию!")
```
